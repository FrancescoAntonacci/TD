# Task 6

Il codice riportato nel notebook *sweepfreq.ipynb* risolve il primo problema riportato usando un *logspace* generato tramite numpy, il secondo modulando il campionamento e andando ad osservare che se acquisiamo il nostro segnale che ha frequenza *ff*, allora il periodo di tempo misurato sarà pari a
$$
    T = \frac{\text{n}_\text{per}}{ff}
$$
dove *nper* è una variabile che specifica il numero di periodi che vengono considerati dal programma. In questa maniera abbiamo che se vogliamo campionare il segnale utilizzando una frequenza di sampling $f_s$ allora abiamo bisogno di ben $f_s \frac{\text{nper}}{ff}$ punti: tale numero, tuttavia, non deve superare 8192 (la dimensione massima a cui facciamo lavorare il buffer di memoria delle misure), pertanto va a calcolare la "frazione" a cui AD2 va a campionare in maniera tale che rispetti la relazione dividendolo per *npt*: in questa maniera risolve il problema prendendo
$$
    df = \left\lceil \frac{f_s \text{n}_\text{per}}{\text{npt} * ff} * 100 \cdot 10^6 \right\rceil,
$$
ovvero la frazione della frequenza massima a cui AD2 campiona coprendo tutto l'intervallo con 8192, approssimandolo verso l'alto proprio per evitare di andare oltre il numero di punti, usando quest'ultimo per calcolare il *df*. <br />
Il terzo e ultimo problema sulle ampiezze negative viene risolto controllando che il valore stimato dal fit non sia negativo: nel caso in cui non sia così, il programma inverte il segno dell'ampiezza e aggiunge alla fase un angolo $\pi$ per far quadrare le cose.


# Task 7


E' stato modificato il codice *sweepfreq.ipynb* per ottenere valor medio e la deviazione standard campione iterando la misura dei segnali e il calcolo del guadagno 30 volte.
Sono stati acquisiti i dati e si è fatto il best-fit. <br />
Sappiamo che l'espressione analitica del guadagno $\mathcal{H}(\omega)$ è data da
$$
    \mathcal{H}(\omega) = \frac{j\omega \tau_1}{(1+j \omega \tau_1)(1 + j \omega \tau_2) + j \omega R_1 C_2} = \frac{j\omega \tau_1}{(1+j\omega \tau_A)(1+j \omega \tau_B)} 
$$

Risulta, effettuando tutto il conto, che
$$\begin{align*}
&\tau_{\text{A},\text{B}} = \frac{\tau_1 + \tau_2 + R_1 C_2 \pm \sqrt{(\tau_1 + \tau_2 + R_1 C_2)^2 - 4 \tau_1 \tau_2}}{2} \implies \\
&\implies f_{\text{A}, \text{B}} = \frac{2}{\tau_1 + \tau_2 + R_1 C_2 \pm \sqrt{(\tau_1 + \tau_2 + R_1 C_2)^2 - 4 \tau_1 \tau_2}}
\end{align*}$$
Con le misure effettuate col multimetro abbiamo che
$$
    \tau_{A, \text{teor}} = (9.99 \pm 0.01) m\text{s} \\
    \tau_{B, \text{teor}} = (4.67 \pm 0.02) \mu\text{s} \\
    f_{A, \text{teor}} = (68 \pm 2) \, \text{Hz}, \\
    f_{B, \text{teor}} = (31 \pm 1) \, \text{kHz}.
$$
dove abbiamo propagato con gli errori presi dal manuale del multimetro. Usando il modello
$$
    |\mathcal{H}(\omega)| = \left| \frac{2 \pi fj \tau_1}{(1+2\pi j \frac{f}{f_A})(1+2\pi j \frac{f}{f_B})} \right|
$$
abbiamo stimato $\tau_1$, $\tau_2$, $R_1$ e $C_2$, che risultano essere pari a:
$$
    \tau_1 = (9.1490 \pm 0.0004) \text{ms} \\
    \tau_2 = (46.781 \pm 0.004) \mu\text{s} \\
$$
che non risultano compatibili con i valori attesi entro le barre di errore. Sempre tramite fit abbiamo stimato *f_A* e *f_B* e si ha che
$$
    f_A^\star = (74.059 \pm 0.004) \, \text{Hz} \\
    f_B^\star = (31.547 \pm 0.003) \, \text{kHz},
$$
dunque solo la seconda misura non è compatibile con quella teorica prevista. <br /> Il $\chi^2$ ridotto risulta essere pari a $6.8$ il che potrebbe far pensare che ci sia una sottostima degli errori nel modello. Abbiamo, quindi, ripetuto l'analisi stimando l'ampiezza del segnale tramite differenza $V_\text{in}$ e $V_\text{out}$ da cui risulta che il $\chi^2$ ridotto è pari a $0.6$ e i valori stimati sono
$$
    \tau_1^{\star \star} = (9.182 \pm 0.005) \, \text{ms} \\
    \tau_2^{\star \star} = (44.1 \pm 0.1) \, \mu\text{s} \\
$$
e in questa maniera abbiamo che
$$
    f_A^{\star \star} = (73.82 \pm 0.05) \, \text{Hz} \\
    f_B^{\star \star} = (33.37 \pm 0.09) \text{kHz}
$$
ed entrambi i valori risultano non compatibili con quelli attesi con i valori di $R_1$, $C_1$, $R_2$ e $C_2$ misurati col multimetro.
Il guadagno a centro banda risulta essere entro $3 dB$ dal guadagno massimo.
Si sono considerati gli errori non assoluti in quanto i residui risultavano correlati statisticamente.

Confrontando i grafici dei residui dei due fit (*Misure_CRRC_residui.pdf* e *BodePassaBanda_Residuals_maxmethod.pdf*) si osserva:
* estrema somiglianza nelle regioni in cui il modello si discosta di più dai dati sperimentali;
* correlazione statistica dei residui simile.

Essendo le due analisi fatte in momenti diversi e con dati diversi, si deduce la presenza di un'errore sistematico che, per la correlazione dei residui , è probabile sia fortemente legato alla non idealità di AD2 come voltmetro generatore di tensione.

La forte correlazione dei residui esclude che il contributo più grande all'errore sistematico che si osserva sia legato alla tecnica con cui si stima l'ampiezza: se questo fosse stato il fattore principale la correlazione dei residui sarebbe cambiata dipendentemente dal metodo.

Si procede col campionamento e successivamente con effettuare di seganli campionati volontariamente più o meno bene.

```python
import tdwf
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt 
import numpy as np

# -[Configurazione Analog Discovery 2]-----------------------------------------
#   1. Connessiene con AD2
ad2 = tdwf.AD2()
ad2.vdd = 5
ad2.vss = -5
ad2.power(True)
#   2. Configurazione generatore di funzioni
wgen = tdwf.WaveGen(ad2.hdwf)
wgen.w1.ampl = 1
wgen.w1.freq = 10e3
wgen.w1.offs = 0.1
wgen.w1.func = tdwf.funcTriangle
wgen.w1.duty = 100
wgen.w1.start()
#   3. Configurazione oscilloscopio
scope = tdwf.Scope(ad2.hdwf)
scope.fs=10e3
scope.npt=8192
scope.ch1.rng = 8
scope.ch2.rng = 8
scope.trig(True, level = 0.1, hist = 0.1, sour = tdwf.trigsrcCh1)
#   4. Configurazione powersupply
ad2.vdd = 5
ad2.vss = -5
ad2.power(True)

# -[Funzioni di gestione eventi]-----------------------------------------------
def on_close(event):
    global flag_run
    flag_run = False
def on_key(event):
    global flag_run
    global flag_acq
    if event.key == 'x':  # => export su file
        filename = input("Esporta dati su file: ")
        # data = np.column_stack((scope.time.vals, scope.ch1.vals, scope.ch2.vals))
        data = np.column_stack((scope.time.vals, scope.ch1.vals))
        if scope.npt > 8192:
            # info =  f"Acquisizione Analog Discovery 2 - Lunga durata\ntime\tch1\tch2"
            info =  f"Acquisizione Analog Discovery 2 - Lunga durata\ntime\tch1"
        else:
            # info =  f"Acquisizione Analog Discovery 2\nTimestamp {scope.time.t0}\ntime\tch1\tch2"
            info =  f"Acquisizione Analog Discovery 2\nTimestamp {scope.time.t0}\ntime\tch1"
        np.savetxt(filename, data, delimiter='\t', header=info)
    if event.key == ' ':  # => run/pausa misura
        flag_acq = not flag_acq
    if event.key == 'escape':  # => esci dalla misura
        flag_run = False

# -[Ciclo di misura]-----------------------------------------------------------
fig, ax = plt.subplots(figsize=(12,6))
fig.canvas.mpl_connect("close_event", on_close)
fig.canvas.mpl_connect('key_press_event', on_key)
flag_run = True
flag_acq = True
flag_first = True
while flag_run:
    if flag_acq: # l'acquisizione è attiva?
        scope.sample()
    # Visualizzazione
    if flag_first:
        flag_first = False
        hp1, = plt.plot(scope.time.vals, scope.ch1.vals, "-", label="Ch1", color="tab:orange")
        # hp2, = plt.plot(scope.time.vals, scope.ch2.vals, "-", label="Ch2", color="tab:blue")
        plt.legend()
        plt.grid(True)
        plt.xlabel("Time [msec]", fontsize=15)
        plt.ylabel("Signal [V]", fontsize=15)
        plt.title("User interaction: x|space|escape")
        plt.tight_layout()
        plt.show(block = False)
    else:
        hp1.set_ydata(scope.ch1.vals)
        # hp2.set_ydata(scope.ch2.vals)
        fig.canvas.draw()
        fig.canvas.flush_events()

plt.close(fig)
ad2.close()
```
